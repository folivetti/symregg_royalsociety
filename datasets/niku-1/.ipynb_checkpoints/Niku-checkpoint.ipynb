```python
import numpy as np
import pandas as pd
from pyeggp import PyEGGP
from collections import defaultdict
```

```python
df = pd.read_csv("../nikuradse_1.csv")
xs = []
ys = []
rs = []
for i,g in enumerate(df.groupby("r_k")):
    rs.append(g[1].r_k.values[0])
    xs.append(g[1].log_Re.values)
    ys.append(g[1].target.values)
```

```python
reg = PyEGGP(maxSize=8,nonterminals='add,sub,mul,div,log,exp,sqrt,recip,power')
reg.fit_mvsr(xs,ys)
reg.results
```

```python
ix = 4
models = reg.results[reg.results.id==ix]

chosen = models.Numpy.values[-1]
thetas = defaultdict(list)
regs = {}
for t in models.theta.values:
    tv = [float(x) for x in t.split(";")]
    for i, x in enumerate(tv):
        thetas[i].append(x)
for k, v in thetas.items():
    regs[k] = PyEGGP(maxSize=6,nonterminals='add,sub,mul,div,log,exp,sqrt,recip,power')
    regs[k].fit(np.array(rs), np.array(v))
    print(regs[k].results.iloc[-1].loss)

```

```python
def eval_mse(rs,xs,ys):
    y_pred = []
    y_true = []
    for r, x, y in zip(rs,xs,ys):
        t = np.array([regs[j].predict(np.array([r])) for j in range(len(regs))])
        x = x.reshape(-1,1)
        yh = eval(chosen)
        y_pred += list(yh)
        y_true += list(y)
    return np.square(np.array(y_pred) - np.array(y_true)).mean()

eval_mse(rs,xs,ys)
```

```python
print(regs[0].results.iloc[-1].Expression)
print(regs[1].results.iloc[-1].Expression)
print(models.iloc[-1].Numpy)
```

```python
def niku_1(x):
    x0 = x[:,0]
    t0 = np.log((43.820120965977694 / (x0 - 11.604807341082092)))
    t1 = (1.977111817884481 - (x0 ** 0.3432687321137361))
    return ((x[:, 1] + t0)) ** (t1 / x[:, 1])

y_pred = niku_1(df.values)
np.square(y_pred - df.values[:,-1]).mean()
```
